The sunrise/sunset data is not in a format that is easy to work with. In this section, we'll clean up the formatting of this data.

First, re-arrange the columns so they are easier to work with.

```python
sun = pd.read_csv(path+'sunset_sunrise.csv',header=0) #import
sun = sun.dropna(axis=1,how='all') #remove empty columns
sun = sun.set_index('Column1') #Make day the index
sun.columns = ['1_Sunrise','1_Sunset','2_Sunrise','2_Sunset','3_Sunrise','3_Sunset','4_Sunrise'
               ,'4_Sunset','5_Sunrise','5_Sunset','6_Sunrise','6_Sunset','7_Sunrise','7_Sunset',
              '8_Sunrise','8_Sunset','9_Sunrise','9_Sunset','10_Sunrise','10_Sunset','11_Sunrise','11_Sunset',
              '12_Sunrise','12_Sunset'] #Name columns
sun = sun.stack().to_frame().reset_index() #Turn columns into rows
sun['Month'], sun['Type'] = sun['level_1'].str.split('_', 1).str #Create fields for month and type (sunset, sunrise)
sun = sun.drop('level_1',axis=1)
sun.columns = ['Day','Time','Month','Type'] #Rename columns
```

Next, reformat the datetime fields so that it can be joined with the police and weather data.

```python
sun['Time'] = sun['Time'].astype(str).str[:-2] #Reformat time into time type format
sun['Time'] = sun['Time'].astype(str).str[:-2] + ':' + sun['Time'].astype(str).str[-2:]
sun['Year']  = 2012 #insert placeholder year
sun['Hour']  = pd.to_datetime(sun['Time']).dt.hour
sun['Minute']  = pd.to_datetime(sun['Time']).dt.minute
sun['Time']  = pd.to_datetime(sun['Time']).dt.time
sun['DateTime']  = pd.to_datetime(sun[['Year','Day','Month','Hour','Minute']]) #Hour, Minute
sun['Month'] = sun['Month'].astype('int64') #convert types to int
sun['Day'] = sun['Day'].astype('int64') #convert types to int
```

Then, we must generate a sunset/sunrise time for each year between 2012 and 2017. For simplicity, we will assume times are the same for each year.

```python
sun_final = sun.copy(deep=True)
for i in range(0,5):
    sun['Year'] = 2013+i
    sun['DateTime']  = pd.to_datetime(sun[['Year','Day','Month','Hour','Minute']])
    sun_final = pd.concat([sun_final,sun])
```

Last, we need to adjust the times for daylights savings time (DST), since the sunset/sunrise times are always assumed to be in Pacific Standard Time. To obtain the DST date ranges below, I Googled daylights savings times for 2012-2017.

```python
sun_final['DST'] = np.where((sun_final['DateTime'] > '2012-03-11') & (sun_final['DateTime'] < '2012-11-04'),1,0)
sun_final['DST'] = np.where((sun_final['DateTime'] > '2013-03-10') & (sun_final['DateTime'] < '2013-11-03'),1,0)
sun_final['DST'] = np.where((sun_final['DateTime'] > '2014-03-09') & (sun_final['DateTime'] < '2014-11-02'),1,0)
sun_final['DST'] = np.where((sun_final['DateTime'] > '2015-03-08') & (sun_final['DateTime'] < '2015-11-01'),1,0)
sun_final['DST'] = np.where((sun_final['DateTime'] > '2016-03-13') & (sun_final['DateTime'] < '2016-11-06'),1,0)
sun_final['DST'] = np.where((sun_final['DateTime'] > '2017-03-12') & (sun_final['DateTime'] < '2017-11-05'),1,0)

sun_final['Hour'] = sun_final['Hour'].astype('int64') + sun_final['DST']
sun_final['Time']  = pd.to_datetime(sun_final[['Year','Day','Month','Hour','Minute']]).dt.time
sun_final['DateTime']  = pd.to_datetime(sun_final[['Year','Day','Month','Hour','Minute']])
sun_final['Hour']  = pd.to_datetime(sun_final[['Year','Day','Month','Hour','Minute']]).dt.hour
```

Next, we'll filter on only BFI and filter out any fields that are not relevant to our final analysis

Here, we will filter out missing values and see how many rows see will remove from the data. We'll also do some data type conversion to ensure data can be easily worked with in the analysis section. At the end, we will output the top 5 rows to see what our final table will look like.

```python
print("Original dataframe shape:", weather_final.shape)
print("# of rows with missing temperature: ",len(weather_final[weather_final['tmpf'] == 'M']))
weather_final = weather_final[ (weather_final['tmpf'] != 'M') 
                              & (weather_final[' p01i'] != 'M')
                              & (weather_final[' sknt'] != 'M')
             ] #Remove fields with missing values.

weather_final.columns = ['TimeStamp','Temp(F)','Precipitation(in)','WindSpeed(kts)']
weather_final['Temp(F)'] = weather_final['Temp(F)'].astype('float64') #convert types to floats
weather_final['Precipitation(in)'] = weather_final['Precipitation(in)'].astype('float64')
weather_final['WindSpeed(kts)'] = weather_final['WindSpeed(kts)'].astype('float64')
weather_final.head()
```

Next, we'll clean-up datetime and break out by month, day and hour. This will ensure it can be more easily analyzed and joined with the police and sunset/sunrise datasets.

```python
weather_final['TimeStamp'] = pd.to_datetime(weather_final['TimeStamp'])
weather_final['Year'] = weather_final['TimeStamp'].dt.year
weather_final['Month'] = weather_final['TimeStamp'].dt.month
weather_final['Day'] = weather_final['TimeStamp'].dt.day
weather_final['Hour'] = weather_final['TimeStamp'].dt.hour
weather_final.head(2)
```

Lastly, we'll aggregate the data by year, month, day and hour to create an hourly dataset.

```python
weather_final_hour = weather_final.groupby(['Year','Month','Day','Hour']).agg({'Temp(F)': np.mean,
                                                          'Precipitation(in)': np.mean,
                                                          'WindSpeed(kts)': np.mean}).reset_index()
```

Then, we'll aggregate by just year, month and day to create a daily dataset.

```python
weather_final_day = weather_final_hour.groupby(['Year','Month','Day']).agg({'Temp(F)': [np.mean, np.max, np.min],
                                                          'Precipitation(in)': np.sum,
                                                          'Hour' : np.size,
                                                          'WindSpeed(kts)': np.mean}).reset_index()
weather_final_day.columns = weather_final_day.columns.get_level_values(0)
weather_final_day.columns = ['Year','Month','Day','Temp(F) Mean','Temp(F) High'
                             ,'Temp(F) Low','Precipitation(in)','CountOfObs','WindSpeed(kts)']
```

Next, we'll look at how missing values are affecting our data

```python
print("mean # of observations:",np.mean(weather_final_day['CountOfObs']))
weather_final_day.groupby(weather_final_day['CountOfObs']).size().to_frame() 
```

We see from the above that most days have 23 or more observations, so we shouldn't be highlighly concerned with missing data impacting our results. 

To reduce the risk the risk of any day with a low number of observations impacting our temperature average (and therefore possibly affecting our analysis results), we'll include only days with 22 or more observations.

```python
weather_final_day = weather_final_day[weather_final_day['CountOfObs'] >= 22]
```

## Section C: Negative binomial regression on data

Now let's explore if we should run a linear regression or poisson regression. To see what we should do, we need to know if the assumptions of these are models are met. 

Let's look first to see if crime follows a poisson distribution.

```python
plt.hist(wc_day[[typ]].as_matrix(),bins='auto')
plt.show()
plt.hist(wc_day[[typ]].as_matrix(),bins=range(0,100))
plt.show()
plt.hist(wc_hour_ss[[typ]].as_matrix(),bins=range(0,50))
plt.show()
```

```python
print("Mean: ",wc_day[['Flag_ViolentCrime']].mean())
print("Variance: ",wc_day[['Flag_ViolentCrime']].var())
print("Standard deviation: ",wc_day[['Flag_ViolentCrime']].std())
```

```python
print("Mean: ",wc_hour_ss[['Flag_ViolentCrime']].mean())
print("Variance: ",wc_hour_ss[['Flag_ViolentCrime']].var())
print("Standard deviation: ",wc_hour_ss[['Flag_ViolentCrime']].std())
```

The data appears to follow a poisson distribution at least loosely, however our mean is not anywhere close to equal to our variance thus we cannot use poisson regression model. Let's try negative binomial regression instead, since this is generally recognized as an alternative regression on count data where the strict poisson assumption of mean does equal to variance is not met.

We'll run the regression on both the hourly and daily datasets.

Based on previous analysis, we will include temperature as well as month, weekday and weekday/hour as features in our model. We'll also add in daylight as a categorical value since we might hypothesize more crime occurs generally when it's dark.

Even though the daily high is 97% correlated with mean temperature, we'll add it as a feature with an interaction term with temperature mean.

```python
formula_day = typ + ' ~ Temp_F_Mean + Temp_F_High + C(Month) + C(Weekday) + Temp_F_High*Temp_F_Mean'
formula_hr = typ + ' ~ Temp_F + C(Month) + C(Weekday) + C(Weekday_Hour) + C(Daylight)'
```

```python
mod_nb_day = smf.glm(formula=formula_day, data=wc_day, family=sm.families.NegativeBinomial()).fit()
mod_nb_day.summary()
```

```python
c = pd.concat([mod_nb_day.fittedvalues.to_frame(),wc_day[['Flag_ViolentCrime']]],axis=1)
plt.hist(c,bins=range(10,110))
plt.title('Daily NB Model')
plt.legend(['fit', 'actual'])
plt.show()
```

The deviance and pearson chi-squared values divided by our degrees of freedom are roughly 1. This indicates the negative binomial model is a reasonably good fit of our data. Looking at the graph, we see that it will generally tend to overestimate the number of crimes on average and it will miss outlier events.

The daily model indiciates a statistically significant negative relationship between temperature and crime, although it is very slight. Weekday (particularly weekends) and summer months generally seem to be more positively correlated with crime.

```python
mod_nb_hour = smf.glm(formula=formula_hr, data=wc_hour_2, family=sm.families.NegativeBinomial()).fit()
mod_nb_hour.summary()
```

```python
c = pd.concat([mod_nb_hour.fittedvalues.to_frame(),wc_hour_2[['Flag_ViolentCrime']]],axis=1)
plt.hist(c,bins=range(1,50))
plt.title('Hourly NB Model')
plt.legend(['fit', 'actual'])
plt.show()
```

The deviance divided by  degrees of freedom is roughly 1, however the pearson chi-squared values indicates our model is underdispersed. This indicates the negative binomial model is not a great, which is why we have scaled it here. Looking at the plot, we see that it will generally tend to undestimate the number of crimes on average and it will miss outlier events.

This model indicate a statistically significant negative relationship between temperature and violent crime at the hourly level. 

The hourly model also indicates a strong positive relationship between crime on weekend mornings (midnight to 2am), consistent with our heatmap. This relationship is about 80X stronger than the negative relationship between temperature and crime. Ignoring time of day and weekday and looking at night or twilight alone, we see a slightly negative relationship between darkness and crime. 

In conclusion, these models are not a terrific fit for the data but do suggest a slight negative relationship between temperature and crime. They also validate previous observations such as the relationship between weekend evenings and violent crime.

## Section D: Linear regression on data

Even though linear regression analysis does not typically work well on count data, other studies such as the Chicago Tribune studies have observed a linear relationship between crime and temperature. Let's see if we can repeat these findings our data.

We will just look at the daily data here since we know there is an underlying non-linear relationship between time of day and temperature.

```python
plt.scatter(wc_day[['Temp_F_Mean']].as_matrix(), wc_day[[typ]].as_matrix())
plt.xlabel('Mean Daily Temp F')
plt.ylabel('Violent Crimes Per Day')
plt.title('Crime vs. Temperature')
plt.show()
```

```python
wc_day_2 = wc_day[wc_day[typ] > 30]
wc_day_2 = wc_day_2[wc_day_2[typ] < 100]
plt.xlabel('Mean Daily Temp F')
plt.ylabel('Violent Crimes Per Day')
plt.title('Crime vs. Temperature (Outliers removed)')
plt.scatter(wc_day_2[['Temp_F_Mean']].as_matrix(), wc_day_2[[typ]].as_matrix())
plt.show()
```

```python
lr_day_all = smf.ols(formula=formula_day,offset=np.log(wc_day['CountOfObs']), data=wc_day).fit()
lr_day_all.summary()
```

Let's also check the distribution of the residuals to make sure they are normally distributed and our model is a reasonable fit.

```python
plt.hist(lr_day_all.resid,bins='auto')
plt.show()
s.graphics.gofplots.qqplot(lr_day_all.resid,line='s')
plt.show()
```

The above R^2 and distribution of residuals indicates that our model is not a very good of the data. Even so, we still find a statistically negative relationship between temperature and crime when controlling for month and weekday. Summer and Fall months and weekends are also related to increases in crime. This relationship is much stronger than the relationship between crime and temperature.

Next, we'll repeat this for the filtered dataset.

```python
lr_day_filt = smf.ols(formula=formula_day,offset=np.log(wc_day_2['CountOfObs']), data=wc_day_2).fit()
lr_day_filt.summary()
```

```python
plt.hist(lr_day_filt.resid,bins='auto')
plt.show()
s.graphics.gofplots.qqplot(lr_day_filt.resid,line='s')
plt.show()
```

The  distribution of residuals indicates that this model is a better fit of our data than the previous one, but it is still a poor fit. Our R^2 is also still low indicating that that the features in the model (temperature, weekday and month) explain very little of daily differences in crime. 

This model doesn't show a statistically significant impact on crime for any of our features.
